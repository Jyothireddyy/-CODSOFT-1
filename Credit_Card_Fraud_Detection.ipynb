```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Sample dataset with a categorical column
data = {
    'feature1': [1, 2, 3],
    'feature2': ['fraud_Reichel Inc', 'other_value', 'another_value'],
    'label': [0, 1, 0]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Separate features and label
x = df[['feature1', 'feature2']]
y = df['label']

# Convert categorical features to numeric using One-Hot Encoding
x_encoded = pd.get_dummies(x)

# Now x_encoded contains only numeric values
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size=0.2, random_state=42)

# Logistic Regression Model
LR = LogisticRegression(solver='liblinear')
LR.fit(x_train, y_train)

# Predict on test set
y_pred = LR.predict(x_test)
print(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load a sample dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split into train and test sets (use a larger test set size to avoid the issue)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize KNN with n_neighbors less than or equal to number of training samples
knn = KNeighborsClassifier(n_neighbors=min(len(x_train), 4))  # or set a smaller n_neighbors manually
knn.fit(x_train, y_train)

# Predict on the test set
predictions = knn.predict(x_test)
print(predictions)
```

```python
predictions = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, jaccard_score, f1_score

# Calculate Accuracy Score (no change needed here)
KNN_Accuracy_Score = accuracy_score(y_test, predictions)

# Calculate Jaccard Index (with an average method for multiclass classification)
KNN_JaccardIndex = jaccard_score(y_test, predictions, average='macro')  # or 'micro', 'weighted'

# Calculate F1 Score (with an average method for multiclass classification)
KNN_F1_Score = f1_score(y_test, predictions, average='macro')  # or 'micro', 'weighted'

# Print results
print(KNN_Accuracy_Score)
print(KNN_JaccardIndex)
print(KNN_F1_Score)
```
